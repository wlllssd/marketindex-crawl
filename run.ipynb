```python
import pandas as pd
import copy
```

```python
total_list = []
raw = ""
company = []
i = 0

f = open("raw.html", "r")
while raw != "EOF":
    raw = f.readline()
    raw = raw.strip('\n')
    company.append(raw)
    if raw == '</tr>':
        temp = copy.deepcopy(company)
        real_data = []
        
        # Rank
        # <td>1</td>
        left = temp[1].find('<td>')+4
        right = temp[1].find('</td>')
        real_data.append(temp[1][left:right])

        # Rank change?(not sure)
        #  -
        if temp[3] == '-':
            real_data.append('no change')
        # <span class="green-up-arrow"></span>
        elif temp[3].find("<span class=\"green-up-arrow\"></span>")!= -1:
            real_data.append('up')
        # <span class="red-down-arrow"></span>
        elif temp[3].find("<span class=\"red-down-arrow\"></span>")!=-1:
            real_data.append('down')
        else:
            real_data.append('error')

        # Code
        # <td><a href="/asx/cba">CBA</a></td>
        left = temp[4].find('>',temp[4].find('<a'))+1
        right = temp[4].find('<',left)
        real_data.append(temp[4][left:right])

        # Company
        # <td>Commonwealth Bank</td>
        left = temp[6].find('<td>')+4
        right = temp[6].find('</td>')
        real_data.append(temp[6][left:right])

        # Price
        # <td>$75.40</td>
        left = temp[7].find('<td>')+4
        right = temp[7].find('</td>')
        real_data.append(temp[7][left:right])

        # Change 
        # +0.23 </td>
        right = temp[9].find('</td>')-1
        real_data.append(temp[9][:right])

        # % Chg
        # +0.31% </td>
        right = temp[11].find('</td>')-1
        real_data.append(temp[11][:right])

        # Mkt Cap
        # <td>$133.48 B</td>
        left = temp[12].find('<td>')+4
        right = temp[12].find('</td>')
        real_data.append(temp[12][left:right])

        # 1 Year
        # +6.23% </td>
        right = temp[14].find('</td>')-1
        real_data.append(temp[14][:right])

        total_list.append(real_data)
        company.clear()
        
f.close()
```

```python
name = ['Rank',' ','Code','Company','Price','Change','% Chg','Mkt Cap','1 Year']

out = pd.DataFrame(data=total_list,columns=name,index=None)
out.head(10)
```

```python
out.to_csv('result.csv')
```
